```python
import pandas as pd
import math
```

```python
d=pd.read_csv('Titanic.csv')
d
```

```python
mean_Age=math.floor(d['Age'].mean())
mean_Age
```

```python
d['Age']=d['Age'].fillna(mean_Age)
d['Age']
```

```python
d.head()
```

```python
df=d.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
df
```

```python
input=df.drop(['Survived'],axis='columns')
input
```

```python
target=df['Survived']
target
```

```python
input.Sex=input.Sex.map({'male':1,'female':2})
input
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(input,target,test_size=0.2,random_state=10)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
model.predict(X_test)
```
